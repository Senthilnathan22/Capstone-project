```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    # x shape will mention the columns&rows that will list in i value and then it will calculate with vif x values then it will be saved as list
    vif["VIF"]=[variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

# Unpaired test

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
median_slag = dataset['slag'].median()
low_slag = dataset[dataset['slag'] <= median_slag]['csMPa']
high_slag = dataset[dataset['slag'] > median_slag]['csMPa']
ttest_ind(low_slag,high_slag)
```

```python
def ttest_on_feature(feature, dataset, target='csMPa'): # Feature=columns need to be tested, dataset=Dataframe, target=output column in default
    dataset = dataset.dropna(subset=[feature, target])
#here we use median value form feature so that we can use this value to split data into 2 groups as low and high
    median_value = dataset[feature].median()
# All rows where the feature value is less than or equal to the median and takes the value from target coulmn   
    group1 = dataset[dataset[feature] <= median_value][target]
# All rows where the feature value is greater than the median, again taking the target column values.
    group2 = dataset[dataset[feature] > median_value][target]
    
    t_stat, p_val = ttest_ind(group1, group2)
    
    print(f"Group 1 (<= median {median_value}) mean {target}: {group1.mean():.2f}") # Shows the average csMPa (target) for both groups
    print(f"Group 2 (> median {median_value}) mean {target}: {group2.mean():.2f}")
    print(f"T-statistic: {t_stat:.4f}") # Prints the results of the t-test: the t-statistic and the p-value rounded to 4 decimal places
    print(f"P-value: {p_val:.4f}")
    
    if p_val < 0.05:
        print("\U0001F449 Statistically significant difference in target!")
    else:
        print("\U0001F449 No significant difference in target.")

    return t_stat, p_val
```

```python
ttest_on_feature('superplasticizer', dataset)
```

# Paired T test

```python
def paired_ttest(feature1, feature2, dataset):
    from scipy.stats import ttest_rel
    paired_data = dataset.dropna(subset=[feature1, feature2])
    t_stat, p_val = ttest_rel(paired_data[feature1], paired_data[feature2])

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    if p_val < 0.05:
        print("ðŸ‘‰ Statistically significant difference between the two features.")
    else:
        print("ðŸ‘‰ No significant difference between the two features.")
    return t_stat, p_val
```

```python
paired_ttest("slag", "flyash", dataset)
```

### This data has only Numerical value and no categorical value however am doing on my self interest.
### we can simulate one-way ANOVA by turning a numerical feature into categories

```python
from scipy.stats import f_oneway
import pandas as pd

def one_way_anova(feature, dataset, target='csMPa', bins=3, labels=None):
    dataset = dataset.dropna(subset=[feature, target])
    
    if labels is None:
        labels = [f"Group_{i+1}" for i in range(bins)]
    
# Create binned groups from the feature,
#Qcut Take a numeric column and split it into q equal parts based on the dataâ€™s distribution not the values themselves.
    dataset[f'{feature}_category'] = pd.qcut(dataset[feature], q=bins, labels=labels, duplicates='drop')
    
    # Extract target values for each group
    groups = [group[target].values for name, group in dataset.groupby(f'{feature}_category')]
    f_stat, p_val = f_oneway(*groups)
    
    print(f"\nOne-Way ANOVA: {target} ~ {feature}_category")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    if p_val < 0.05:
        print("ðŸ‘‰ Statistically significant difference between groups.")
    else:
        print("ðŸ‘‰ No significant difference between groups.")
    
    return f_stat, p_val
```

```python
one_way_anova('slag', dataset, target='csMPa', bins=2, labels=['Low','High'])
```

#### Same for two way

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

def two_way_anova(feature1, feature2, dataset, target='csMPa', bins=3):
    dataset = dataset.dropna(subset=[feature1, feature2, target])

    # Create binned categories for both features
    dataset[f'{feature1}_cat'] = pd.qcut(dataset[feature1], q=bins, labels=[f'{feature1}_L{i+1}' for i in range(bins)], duplicates='drop')
    dataset[f'{feature2}_cat'] = pd.qcut(dataset[feature2], q=bins, labels=[f'{feature2}_L{i+1}' for i in range(bins)], duplicates='drop')

    # Formula for Two-Way ANOVA with interaction
    formula = f'{target} ~ C({feature1}_cat) + C({feature2}_cat) + C({feature1}_cat):C({feature2}_cat)'

    model = ols(formula, data=dataset).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    print(f"\nTwo-Way ANOVA: {target} ~ {feature1} + {feature2}")
    print(anova_table)

    return anova_table
```

```python
two_way_anova('cement', 'water', dataset, target='csMPa', bins=3)
```
